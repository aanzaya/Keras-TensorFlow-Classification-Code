## Train Test Split

```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```


## Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Creating the Model

    # For a binary classification problem

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))

model.add(Dense(units=1,activation='sigmoid')) # Binary Clasification

model.compile(loss='binary_crossentropy', optimizer='adam')
```

## Training the Model 

### Example One: Choosing too many epochs and overfitting!

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data= (X_test,y_test))
```

### Analyze and Plot the loss

```python
model.history.history
```

```python
pd.DataFrame(model.history.history)
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

The model shows both losses decreasing but at a certain point the training loss keeps going down while the validation loss begins to increase meaning we are overfitting the model because of training too many epochs

## Example Two: Early Stopping

We obviously trained too much! Let's use early stopping to track the val_loss and stop training once it begins increasing too much!

### We copy the model to prevent us from traing the first model

```python
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

Stop training when a monitored quantity has stopped improving.

Arguments:
    monitor: Quantity to be monitored.
    min_delta: Minimum change in the monitored quantity
        to qualify as an improvement, i.e. an absolute
        change of less than min_delta, will count as no
        improvement.
    patience: Number of epochs with no improvement
        after which training will be stopped.
    verbose: verbosity mode.
    mode: One of `{"auto", "min", "max"}`. In `min` mode,
        training will stop when the quantity
        monitored has stopped decreasing; in `max`
        mode it will stop when the quantity
        monitored has stopped increasing; in `auto`
        mode, the direction is automatically inferred
        from the name of the monitored quantity.

```python
from tensorflow.keras.callbacks import EarlyStopping
```

Mode: For Accuracy mode=max; For Loss mode=min

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25) 
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data= (X_test,y_test),callbacks=[early_stop])
```

### Analyze and Plot the loss

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

Both models have flatten out and no increase in validation loss. Where they start to spread out should be when we should stop training the model

## Example Three: Adding in DropOut Layers

```python
from tensorflow.keras.layers import Dropout
```

### We copy the model to prevent us from traing the second model (Including the dropout)

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5)) # Dropout

model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

Dropout (0.5) means: Half on the neurons during each batch of epochs of 30 and 15 (as above) layers that we are training will be turned off and their weight and biases will not be updated

### Combining with early stopping

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data= (X_test,y_test),callbacks=[early_stop])
```

### Analyze and Plot the loss

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

This is exactly the kind of model we want. The losses are quickly going down and they essentially flattening out at the same rate
